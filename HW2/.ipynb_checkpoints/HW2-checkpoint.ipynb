```python
train.loc[train["Attrition"] == "No", "Attrition"] = 0.0
train.loc[train["Attrition"] == "Yes", "Attrition"] = 1.0
```

```python
cat_cols = list(train)
for col in cat_cols:
    train[col] = pd.Categorical(train[col]).codes
```

```python
# def binize(train_set, col_name):
# #     bin_sizes = [10,40,100]
#     for bin_size in bin_sizes:
#         column = train_set[~np.isnan(train_set[col_name])][col_name]
#         bins = np.arange(column.min(axis=0),column.max(axis=0), (column.max(axis=0) - column.min(axis=0))/bin_size)
#         _bin = np.digitize(column, bins)
#         bin_count = np.bincount(_bin)
#         print(bin_size,": ")
#         print(bin_count)
        
#         train_col_no_nan = train_set[~np.isnan(train_set[col_name])][col_name]
#         hist, bin_edge = np.histogram(train_col_no_nan, bin_size)
#         print(col,':',np.sum(bin_edge==0))
#         plt.fill_between(bin_edge.repeat(2)[1:-1],hist.repeat(2),facecolor="steelblue")
#         plt.title(bin_size)
#         plt.show()
        
def binize(train_set, col_name, plot=0):
    do = True
    if(col_name == "Age"):bin_size=10
    elif(col_name == "DistanceFromHome"): bin_size=10
    elif(col_name == "HourlyRate"): bin_size=10
    elif(col_name == "DailyRate"): bin_size=10
    elif(col_name == "MonthlyRate"): bin_size=10
    elif(col_name == "TotalWorkingYears"): bin_size=10
    elif(col_name == "YearsAtCompany"): bin_size=10
    elif(col_name == "YearsInCurrentRole"): bin_size=10
    elif(col_name == "YearsSinceLastPromotion"): bin_size=10
    elif(col_name == "YearsWithCurrManager"): bin_size=10
    else:
        do = False
    _bin = train_set[~np.isnan(train_set[col_name])][col_name]
    
    if(do):
        bins = np.arange(_bin.min(axis=0),_bin.max(axis=0), (_bin.max(axis=0) - _bin.min(axis=0))/bin_size)
        _bin = np.digitize(_bin, bins)
        bin_count = np.bincount(_bin)
        train_col_no_nan = train_set[~np.isnan(train_set[col_name])][col_name]
        hist, bin_edge = np.histogram(train_col_no_nan, bin_size)
        for i in range (_bin.shape[0]):
            _bin[i] = bins[_bin[i]-1]
        if(plot == 1):
            plt.fill_between(bin_edge.repeat(2)[1:-1],hist.repeat(2),facecolor="steelblue")
            plt.title(col_name)
            plt.show()
    return _bin,bin_count
        
```
